```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/drive/MyDrive/dataset/extra_3.csv')
df.head(50)
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Data Import and Preprocessing


# Checking the first few rows of the dataset to understand the structure
print(df.head())

# Clean the data - handling missing values (fill or drop)
# Filling missing values with mean or median (depending on context)
df.fillna(df.mean(), inplace=True)  # Or use df.dropna() to remove rows with missing values

# Ensure no negative values for resource usage
df = df[df['catalyst_quantity'] >= 0]
df = df[df['energy_consumption'] >= 0]
df = df[df['raw_material_input'] >= 0]

# Step 3: Resource Utilization Calculation
# Assuming column names: 'catalyst_quantity', 'energy_consumption', 'raw_material_input', 'biofuel_yield'
df['total_resource_input'] = df['catalyst_quantity'] + df['energy_consumption'] + df['raw_material_input']

# Step 4: Efficiency Analysis
# Resource efficiency is biofuel yield divided by total resource input
df['efficiency'] = df['biofuel_yield'] / df['total_resource_input']

# Step 5: Visualization
# Bar chart for catalyst usage, energy consumption, and raw material input
df.plot(kind='bar', x='production_run', y=['catalyst_quantity', 'energy_consumption', 'raw_material_input'],
        figsize=(10, 6), title='Resource Utilization per Production Run')

plt.xlabel('Production Run')
plt.ylabel('Resource Usage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot showing the relationship between resource input and biofuel yield
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['total_resource_input'], y=df['biofuel_yield'], color='blue')
plt.title('Relationship Between Resource Input and Biofuel Yield')
plt.xlabel('Total Resource Input')
plt.ylabel('Biofuel Yield')
plt.tight_layout()
plt.show()

# Scatter plot for resource efficiency vs production run
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['production_run'], y=df['efficiency'], color='green')
plt.title('Resource Efficiency per Production Run')
plt.xlabel('Production Run')
plt.ylabel('Efficiency (Yield/Input)')
plt.tight_layout()
plt.show()

# Optional: Save the cleaned and processed data for further use
df.to_csv('processed_biofuel_data.csv', index=False)


```
